```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# importiamo il dataset titanic

df = pd.read_csv('train.csv')
df.head()
```

```python
# plottare una distribuzione dei passeggeri
plt.figure(figsize=(12,6))

sns.distplot(df["Age"], hist=True)
plt.xticks(np.linspace(0, 100, 20))
plt.show()
```

```python
# barplot
plt.figure(figsize=(12,6))

#sns.barplot(x="Sex", y="Survived", data=df)     # i parametri da inserire saranno i label delle due colonne su cui calcolare le percentuali e il dataframe di origine
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=df)   # il terzo parametro hue è il nome della terza colonna su cui parametrizzare
plt.show()
```

```python
# plottare una regressione
plt.figure(figsize=(10,8))
sns.regplot(x="Age", y="Fare", data=df)
plt.plot()
```
