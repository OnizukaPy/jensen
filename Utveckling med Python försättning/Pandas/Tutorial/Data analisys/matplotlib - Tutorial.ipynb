```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# come si compone il grafico:

# La FIGURE è il contenitore principale del grafico che è chiamato AXES. Ogni FIGURE può contenere più grafici.
# Ogni grafico ha degli assi AXIS un titolo TITLE etc..

# Creare una FIGURE contenitore
fig = plt.figure(
    figsize=(16,6)          # con il parametro figsize possiamo impostare la grandezza del grafico mettendo i due parametri in pollici in una turple
)

ax = plt.axes()
```

```python
# plottiamo ad esempio una linea continua di punti
x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

plt.plot(x, y)          # il metodo plot() crea il grafico 
plt.show()              # il metodo show() lo mostra
```

```python
# se non d'ho l'asse x mi usa automaticamente un np.array([0,1,2, .. etc])
y = np.array([1, 2, 3, 4, 5, 6])
plt.plot(y)
plt.plot(y**2)
plt.plot(y**3)
plt.show()
```

```python
# customizzare il grafico:

y = np.array([1, 2, 3, 4, 5, 6])

plt.plot(y, "green")            # per passare il colore basta aggiungerlo come parametro nel metodo plot().
plt.plot(y**2, "pink")

# Griglia
plt.grid()

# asse x più preciso
plt.xticks(np.linspace(0, 10, 10))

# labels e titolo
plt.xlabel("x")                 # asse x metodo xlabel e stringa con il nome del label
plt.ylabel("y")                 # asse y metodo ylabel e stringa con il nome del label
plt.title("Grafico di base")    # totolo metodo title e stringa con il titolo del grafico




# Legenda
plt.legend(["y", "y2"])         # metodo legend con argomento una lista di etichette in ordine di plotting

plt.show()
```

```python
# come creare più grafici all'interno della stessa figure, ovvero i subplots.

# impostiamo il contenitore con una dimensione di 10x8
plt.figure(figsize=(10, 8))

x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 6, 9, 12, 15])

# usando il metodo subplot() inserendo come parametri: larghezza, altezza e indice del subplot, vado a visionare più grafici nello stesso conteni
# di grandezza predefinita

plt.subplot(2, 2, 1)
plt.plot(x, y)

plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, y**2)

plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x, y**3)

plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x, y**4)

plt.grid()
plt.show()

```

```python
# Alcune funzioni:

# funzioni con linea continua
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x), "blue")
plt.grid()
plt.show()
```

```python
# funzioni con scatter
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.scatter(x, y, color="green")        # nel metodo vanno inseriti i due elenchi di coordinate seguiti dal parametro per il colore dello scatter
plt.grid()
plt.show()
```

```python
# mostrare un istogramma delle frequenze
data1 = np.random.randn(1000)

plt.hist(data1)
plt.show()
```

```python
# grafico a barre
data2 = [5., 25., 50., 20.]

plt.bar(range(len(data2)), data2)       # come parametri serveno il numero di colonne da printare e i dati delle rispettive colonne
plt.show()
```

```python
# grafico a torta
plt.figure(figsize=(7,7))

x = [35, 25, 20 , 20]           # i valori devono essere percentuali la cui somma è 100

labels = ["1", "2", "3", "4"]   # etichette degli spicchi

plt.pie(x, labels=labels)
plt.show()
```

```python
# boxplot

data3 = np.random.randn(100)

plt.boxplot(data3)

plt.show()
```
