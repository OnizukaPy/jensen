```python
corr = df.corr()

fig, ax = plt.subplots(figsize=(16,6))

sns.heatmap(
    corr,
    vmin=-1, vmax=1, center=0, annot=True,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
```
