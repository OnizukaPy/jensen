```python
import pandas as pd
```

```python
# Funktionen läser in en database och returnerar en normaliserad database utan hantera original-databasen.
def normalize(df):
    """
    df är database för att normalisera.
    """

    # Skapa två kopior av df. En för att normalisera och en för att spara resultatet.
    result = df.copy()
    pf = df.copy()

    # Ta bort alla kolumner som inte är numeriska från result.
    result = result[result.dtypes[result.dtypes != 'object'].index]

    # Normalisera alla numeriska kolumner i result.
    for feature_name in result.columns[1:]:

        max_value = result[feature_name].max()
        min_value = result[feature_name].min()
        result[feature_name] = (result[feature_name] - min_value) / (max_value - min_value)
        pf[feature_name] = result[feature_name]

    # Spara resultatet i en csv-fil.
    pf.to_csv('normalized.csv')

    return pf
```

```python
df = pd.read_csv('Automobile_data.csv')
df.head()
```

```python
pf = normalize(df)
pf.head()
```
