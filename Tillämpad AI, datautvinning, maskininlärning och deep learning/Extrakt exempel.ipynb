```python
import pandas as pd
import requests
```

```python
def get_json(url):
    response = requests.get(url)
    return response.json()
```

```python
# Read in the data from swapi.co
url = 'https://swapi.dev/api/'
data = get_json(url)
```

```python
dataset = {}
for key, value in zip(data.keys(), data.values()):
    dataset[key] = get_json(value)
```

```python
df = pd.DataFrame(dataset)
df
```

```python
people = df['people'][3]
planets = df['planets'][3]
films = df['films'][3]
species = df['species'][3]
vehicles = df['vehicles'][3]
starships = df['starships'][3]
```

```python
df_people = pd.DataFrame(people)
df_planets = pd.DataFrame(planets)
df_films = pd.DataFrame(films)
df_species = pd.DataFrame(species)
df_vehicles = pd.DataFrame(vehicles)
df_starships = pd.DataFrame(starships)
```
