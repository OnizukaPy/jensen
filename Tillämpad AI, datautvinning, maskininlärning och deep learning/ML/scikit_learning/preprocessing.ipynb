```python
#Idag: Preprocessing av data
# För att vi ska kunna göra en bra modell behöver vi först förbereda datan.
# Det kan typiskt saknas data, eller finnas kategorisk (strängar) data som
# vill göra om till numeriuska värden. 

import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
# Ladda in lite schysst data
iris = datasets.load_iris()

# Skapa en dataframe
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                        columns= iris['feature_names'] + ['target'])
```

```python
df.head()
```

```python
# Ta bort lite data, och gör vissa rader till kategorisk data

df.loc[3:7, 'sepal length (cm)'] = np.nan
df.loc[8:15, 'target'] = 'setosa'
df.loc[56:64, 'target'] = 'versicolor'
```

```python
df[7:16]
```

```python
df.head()
```

```python
# Hantera saknad data
df["sepal length (cm)"].fillna(df['sepal length (cm)'].mean(), inplace=True)
```

```python
df.head()
```

```python
# Gör om våra kategoriska värden till dummy värden

df = pd.get_dummies(df,columns=["target"], prefix = "prox")
```
