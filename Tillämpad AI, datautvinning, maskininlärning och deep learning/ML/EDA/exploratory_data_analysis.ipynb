```python
# Idag: Exploratory Data Analysis
# Vi vill undersöka vårt dataset, och se hur de olika featuresen förhåller sig till varandra
# Vi kommer att använda oss av pandas, matplotlib, sklearn och seaborn(?)

import pandas as pd
from sklearn import datasets
import numpy as np
```

```python
# Importera datan

iris = datasets.load_iris()
```

```python
# Vi inspekterade datasetet genom att klicka på texten "in a text editor"
iris
```

```python
# Gör datan till en dataframe
# Från början hade vi i dataparameter en lista med iris["data"], samt iris["target"]
# Då fick vi 4 kolumner med värdena från data, och headers från feature_names
# samt en kolumn "target" med värdena från "target" i irisdatasetet.
# För snyggare plottar, så ändrade vi kolumnnamnet i dataframen från "target"
# till "species", och ändrade värdena från de i "target", till den faktiska arten.
# Detta gjorde vi genom att casta "target" till ["target_names"]["target"]m, vilket
# gjorde värdena i "target", 0, 1, 2, till motsvarande plats i "target_names".
df = pd.DataFrame(data= np.c_[iris['data'], iris["target_names"][iris['target']]],
                        columns= iris['feature_names'] + ['species'])
```

```python
# Casta om kolumnerna till float64

df["sepal width (cm)"] = pd.to_numeric(df["sepal width (cm)"])
df["sepal length (cm)"] = pd.to_numeric(df["sepal length (cm)"])
df["petal width (cm)"] = pd.to_numeric(df["petal width (cm)"])
df["petal length (cm)"] = pd.to_numeric(df["petal length (cm)"])
```

```python
# Inspektera så att den ser bra ut
df
```

```python
# Äntligen lite Exploratory Data Analysis
# Vi börjar med att använda lite pandasfunktioner
# Här lär vi oss om dataframens nollvärden, samt vilka datatyper kolumnerna har

df.info()
```

```python
# En annan grej är att undersöka siffervärden
# Vi kan notera medianen, min, max, standarddeviationen etc. 
# Inga beslut fattas direkt nu, i det här fallet 

df.describe()
```

```python
# Kika efter nullvärden mer explicit

df.isnull().sum()
```

```python
# Inspektera unika samples / duplikat
# KRITISKT: skriv till en ny variabel

dupes = df.drop_duplicates(subset="species")
dupes
```

```python
# Vi kan inspektera unika targets
# Om det är många fler av en sort, så kan det behöva hanteras
# Detta gör vi i en senare vecka.
df.value_counts("species")
```

Nästa steg: Visualisering
- Vi inspekterar vår data, och gör relevanta plottar.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Inspekterar mängden av arter i datasetet
# Med seaborn så lägger vi in datan separat, och sedan vad från dfn som vi vill plotta.
# Om det är stora skillnader i mängd på våra labels/targets/outputs, så kan vi notera det

sns.countplot(x = "species", data = df)
plt.show()
```

```python
# Vi kan ställa upp en scatterplot, där vi jämför egenskaper hos olika features
# mellan olika arter. 

sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", hue="species", data=df)
plt.legend(bbox_to_anchor=(1,1), loc = 2)
plt.show()
```

```python
# Vi vill göra exakt samma sak, men jämföra andra features
# Vi ser här att det verkar ganska enkelt att separera de olika arterna
# i den här dimensionen. 

sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue="species", data=df)
plt.legend(bbox_to_anchor=(1,1), loc = 2)
plt.show()
```

```python
# Fortsätter, annan dimension

sns.scatterplot(x="sepal length (cm)", y="petal width (cm)", hue="species", data=df)
plt.legend(bbox_to_anchor=(1,1), loc = 2)
plt.show()
```

```python
# Vi vill göra exakt samma sak, men jämföra andra features
# Vi ser här att det verkar ganska enkelt att separera de olika arterna
# i den här dimensionen. 

sns.scatterplot(x="petal length (cm)", y="sepal width (cm)", hue="species", data=df)
plt.legend(bbox_to_anchor=(1,1), loc = 2)
plt.show()
```

Summering av dimensionsplottar: 
- Ganska enkelt att separera setosa xD
- Versicolor och virignica kan gå att se skillnad i petal length vs petal width

```python
# Vi vill inspektera alla möjligheter, hur gör vi det? 
# Ett sätt är att använda oss av seaborns pairplot
# Pairplotten ger oss alla möjligheter för de olika dimensionerna
# På diagonalen så kan vi se distributionen av de olika featuresen per art/target/labels

sns.pairplot(df, hue="species", height=2)
plt.show()
```

Vad ser vi?

- Vi förstår varför det var lätt att urskilja setosa, deras petal length och width är mycket mindre än de andras. 
- Den största setosan är till och med mindre än den minsta av de andra! 

```python
# Mer visualisering!!!!!!!!!!!!!!!!?!?!
# Vi kör lite histogram
# Vi kör lite schyssta subplots!

fig, axes = plt.subplots(2,2, figsize=(10,10))

axes[0,0].set_title("Sepal Length")
axes[0,0].hist(df["sepal length (cm)"], bins = 7)

axes[0,1].set_title("Sepal Width")
axes[0,1].hist(df["sepal width (cm)"], bins = 5)

axes[1,0].set_title("Petal Length")
axes[1,0].hist(df["petal length (cm)"], bins = 6)

axes[1,1].set_title("Petal Width")
axes[1,1].hist(df["petal width (cm)"], bins = 8)
```

```python
# Ett till bra sätt att visualisera distribution av våra variabler

plot = sns.FacetGrid(df, hue="species")
plot.map(sns.distplot, "sepal length (cm)").add_legend()

plot = sns.FacetGrid(df, hue="species")
plot.map(sns.distplot, "sepal width (cm)").add_legend()

plot = sns.FacetGrid(df, hue="species")
plot.map(sns.distplot, "petal length (cm)").add_legend()

plot = sns.FacetGrid(df, hue="species")
plot.map(sns.distplot, "petal width (cm)").add_legend()

plt.show()
```

Summering av histogram:
