```python
# trasformiamo la colonna 'Longevity (yrs)' in una colonna numerica divisa in due colonne 'Longevity (yrs) min' e 'Longevity (yrs) max'
# per fare questo, creiamo una funzione che prende in input una stringa e restituisce una lista di due elementi
# la funzione prende in input una stringa del tipo '12-15' e restituisce una lista di due elementi [12, 15]
def split_string(stringa):
    lista = stringa.split('-')
    lista = [int(x) for x in lista]
    return lista

# applichiamo la funzione alla colonna 'Longevity (yrs)'
# usiamo il metodo apply() di pandas per applicare una funzione alla colonna 'Longevity (yrs)'
df['Longevity (yrs)'] = df['Longevity (yrs)'].apply(split_string)

# creiamo le due nuove colonne 'Longevity (yrs) min' e 'Longevity (yrs) max'
df['Longevity (yrs) min'] = df['Longevity (yrs)'].apply(lambda x: x[0])     # la funzione lambda prende in input una lista e restituisce il primo elemento
df['Longevity (yrs) max'] = df['Longevity (yrs)'].apply(lambda x: x[1])     # la funzione lambda prende in input una lista e restituisce il secondo elemento

# ora possiamo eliminare la colonna 'Longevity (yrs)'
df.drop('Longevity (yrs)', axis=1, inplace=True)  # potremmo anche usare del df['Longevity (yrs)']
```

```python
df.head()   
```

```python
# effettuiamo la stessa cosa per la colonna 'Height (cm)'
df['Height (in)'] = df['Height (in)'].apply(split_string)
df['Height (in) min'] = df['Height (in)'].apply(lambda x: x[0])
df['Height (in) max'] = df['Height (in)'].apply(lambda x: x[1])
df.drop('Height (in)', axis=1, inplace=True)
```

```python
# analiziamo i dati della colonna 'Character Traits' e creiamo una colonna per ogni caratteristica
# la colonna 'Character Traits' contiene una lista di caratteristiche
# creiamo una funzione che prende in input una lista e restituisce una lista di caratteristiche
def split_string_2(stringa):
    lista = stringa.split(',')
    lista = [x.strip() for x in lista]
    return lista

# applichiamo la funzione alla colonna 'Character Traits'
df['Character Traits'] = df['Character Traits'].apply(split_string_2)

# creiamo una lista di caratteristiche
lista_caratteristiche = []
for lista in df['Character Traits']:
    for caratteristica in lista:
        if caratteristica not in lista_caratteristiche:
            lista_caratteristiche.append(str(caratteristica).lower())  

# conserviamo solo i valori unici della lista
import numpy as np
lista_caratteristiche = np.unique(lista_caratteristiche)

# creiamo una colonna per ogni caratteristica
for caratteristica in lista_caratteristiche:
    df[caratteristica] = df['Character Traits'].apply(lambda x: 1 if caratteristica in str(x).lower() else 0)
```

```python
print(lista_caratteristiche)
```

```python
# eliminiamo la colonna 'Character Traits'
df.drop('Character Traits', axis=1, inplace=True)
```

```python
# applichiamo la funzione anche alla colonna 'Common Health Problems'
df['Common Health Problems'] = df['Common Health Problems'].apply(split_string_2)

# creiamo una lista di problemi di salute
lista_problemi_salute = []
for lista in df['Common Health Problems']:
    for problema_salute in lista:
        if problema_salute not in lista_problemi_salute:
            lista_problemi_salute.append(problema_salute)

# conserviamo solo i valori unici della lista
lista_problemi_salute = np.unique(lista_problemi_salute)

# creiamo una colonna per ogni problema di salute
for problema_salute in lista_problemi_salute:
    df[problema_salute] = df['Common Health Problems'].apply(lambda x: 1 if problema_salute in x else 0)

# eliminiamo la colonna 'Common Health Problems'
df.drop('Common Health Problems', axis=1, inplace=True)
```

```python
from libretranslatepy import LibreTranslateAPI

lt = LibreTranslateAPI("https://translate.argosopentech.com/")
```

```python
# creiamo una funzione di traduzione che prende in input una stringa e restituisce la traduzione in italiano
def traduci(stringa):
    # individua la lingua della stringa
    lingua = lt.detect(stringa)[0]['language']
    # traduci la stringa in italiano
    stringa_tradotta = lt.translate(stringa, lingua, 'it')
    return stringa_tradotta
```
